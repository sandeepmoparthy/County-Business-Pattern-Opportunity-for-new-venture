```python
import pandas as pd
import numpy as np
```

```python
stuff= pd.read_csv("texas_data.csv")
stuff1= stuff.loc[stuff.naics.astype(str).str.startswith('62')]
```

```python
small_stuff= stuff1.drop(columns = ["COUNTY_sz_med", "COUNTY_sz_lrg", "state_sz_med", "state_sz_lrg"])
```

```python
# define the data/predictors as the pre-set feature names  
df = pd.DataFrame(small_stuff, columns=["COUNTY_emp","COUNTY_qp1","COUNTY_est"])

# Put the target (housing value -- MEDV) in another DataFrame
target = pd.DataFrame(small_stuff.COUNTY_ap, columns=["COUNTY_ap"])
```

```python
import statsmodels.api as sm

X = df["COUNTY_emp","COUNTY_est"]
y = target["COUNTY_ap"]

# Note the difference in argument order
model = sm.OLS(y, X).fit()
predictions = model.predict(X) # make the predictions by the model

# Print out the statistics
model.summary()
```
